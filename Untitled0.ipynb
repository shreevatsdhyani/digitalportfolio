```python
# Cell 2: Import Libraries
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from groq import Groq
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from fastapi.middleware.cors import CORSMiddleware
import nest_asyncio  # Import nest_asyncio
from pyngrok import ngrok  # Import ngrok
```

```python
# Cell 4: Initialize FastAPI App and CORS
app = FastAPI()

# Allow CORS for frontend communication
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # Allow all origins (update this for production)
    allow_methods=["*"],
    allow_headers=["*"],
)
```

```python
# Cell 5: Load FAISS Vector Store
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

try:
    vector_store = FAISS.load_local(
        "jiopay_index",  # Path to the folder containing index.faiss and index.pkl
        embeddings,
        allow_dangerous_deserialization=True  # Enable deserialization
    )
    print("FAISS vector store loaded successfully!")
except Exception as e:
    print(f"Error loading FAISS vector store: {e}")
```

```python
# Cell 6: Initialize Groq Client
groq_api_key = "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"
client = Groq(api_key=groq_api_key)
```

```python

# Cell 7: Define Query Request Model
class QueryRequest(BaseModel):
    query: str
```

```python

# Cell 8: Create the `/ask` Endpoint
@app.post("/ask")
async def ask(query_request: QueryRequest):
    query = query_request.query

    try:
        # Retrieve relevant documents
        retrieved_docs = vector_store.similarity_search(query, k=3)
        retrieved_texts = "\n\n".join([doc.page_content for doc in retrieved_docs])

        # Generate response using Groq API
        response = client.chat.completions.create(
            model="llama3-8b-8192",
            messages=[
                {"role": "system", "content": "You are a helpful JioPay assistant. Answer only using the provided data."},
                {"role": "user", "content": f"Answer this query based on the provided data:\n\n{retrieved_texts}\n\nQuery: {query}"}
            ]
        )

        # Return response and source references
        return {
            "response": response.choices[0].message.content,
            "sources": [doc.metadata for doc in retrieved_docs]
        }
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
```
